```python
from google.colab import output

def get_url_param(param_name):
    # JavaScript to grab the URL fragment and parse it
    js_code = f"""
    (function() {{
        const params = new URLSearchParams(window.location.hash.substring(1));
        return params.get('{param_name}');
    }})();
    """
    return output.eval_js(js_code)

# Usage
my_value = get_url_param('myParam')
print(f"The value passed is: {my_value}")
```
